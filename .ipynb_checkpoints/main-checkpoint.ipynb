```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import requests
from io import StringIO
from matplotlib.ticker import AutoMinorLocator

colors = ["blue", "green", "red", "cyan", "magenta", "yellow"]

def visualize(data, title):
    df = pd.DataFrame(data)

    df_sorted = df.sort_values("CPU CORES ALLOCATED")

    plt.figure(figsize=(10, 6))
    plt.plot(
        df_sorted["CPU CORES ALLOCATED"], 
        df_sorted["TPS"], 
        marker="o", 
        linestyle="-", 
        label="Transactions per Second",
        linewidth=4,
        markersize=15,
    )
    plt.title(title, fontsize=16)
    plt.xlabel("CPU Cores Allocated", fontsize=14)
    plt.ylabel("Transactions per Second", fontsize=14)
    plt.xticks(df_sorted["CPU CORES ALLOCATED"], fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis="both", linestyle="-", alpha=0.7)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

def n_visualize(data, title):
    plt.figure(figsize=(10, 6))
    for d in data:
        df = pd.DataFrame(d[0])

        df_sorted = df.sort_values("CPU CORES ALLOCATED")

        
        plt.plot(
            df_sorted["CPU CORES ALLOCATED"], 
            df_sorted["TPS"],
            marker="o", 
            linestyle="-", 
            label=d[1],
            linewidth=3,
            markersize=10,
        )
    plt.title(title, fontsize=16)
    plt.xlabel("CPU Cores Allocated", fontsize=14)
    plt.ylabel("Transactions per Second", fontsize=14)
    plt.xticks(df_sorted["CPU CORES ALLOCATED"], fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis="both", linestyle="-", alpha=0.7)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()
```

# E2 Standard

```python
data_e2 = {
    "MACHINE TYPE": [
        "e2-standard-32", 
        "e2-standard-16", 
        "e2-standard-8", 
        "e2-standard-4",
        "e2-standard-2"
    ],
    "CPU TYPE": ["e2", "e2", "e2", "e2", "e2"],
    "CPU CORES ALLOCATED": [32, 16, 8, 4, 2],
    "TPS": [283016.158681, 215930.656338, 109249.522824, 42757.972554, 21473.866045]
}
visualize(data_e2, "TPS vs. CPU Cores Allocated (E2 Standard Machine Type)")
```

# C3D Standard

```python
data_c3d = {
    "MACHINE TYPE": [
        "c3d-standard-16", 
        "c3d-standard-8", 
        "c3d-standard-4"
    ],
    "CPU TYPE": ["c3d", "c3d", "c3d"],
    "CPU CORES ALLOCATED": [16, 8, 4],
    "TPS": [258385.827274, 149961.786033, 75281.430496]
}
visualize(data_c3d, "TPS vs. CPU Cores Allocated (C3D Standard Machine Type)")
```

# C2 Standard

```python
data_c2 = {
    "MACHINE TYPE": [
        "c2-standard-16", 
        "c2-standard-8", 
        "c2-standard-4"
    ],
    "CPU TYPE": ["c2", "c2", "c2"],
    "CPU CORES ALLOCATED": [16, 8, 4],
    "TPS": [247322.520615, 131131.610883, 63878.373262]
}
visualize(data_c2, "TPS vs. CPU Cores Allocated (C2 Standard Machine Type)")
```
